## Data Preprocessing

```python
# Augmenting/Concatenating previous world cup data with latest data

aug_match_data = pd.concat([match_data, recent], axis=0)

aug_match_data = aug_match_data.reset_index(drop = True)
aug_match_data.tail(5)
```

```python
aug_match_data.shape
```

```python
aug_match_data.columns
```

```python
aug_match_data.drop(columns=['Date','Margin', 'Ground'], axis=1, inplace=True)

aug_match_data.head()
```

```python
world_cup_teams = ['India' , 'South Africa' , 'Netherlands' , 'Pakistan', 'New Zealand' ,
                   'Sri Lanka', 'Afghanistan', 'Australia', 'Bangladesh', 'England']
```

```python
df_teams_1 = aug_match_data[aug_match_data['Team_1'].isin(world_cup_teams)]
df_teams_2 = aug_match_data[aug_match_data['Team_2'].isin(world_cup_teams)]

final_teams = pd.concat((df_teams_1, df_teams_2), axis=0)

final_teams.drop_duplicates()

final_teams.count()
```

```python
final_teams = final_teams.reset_index(drop=True)

final_teams.loc[:,'winning_team'] = np.where(final_teams['Winner']==final_teams['Team_1'],0,1)
final_teams.tail()
```

```python
final_teams.drop(columns=['Winner'], axis=1, inplace=True)
final_teams.tail()
```

## Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
# Create dummy variables(one hot encoding)
final_teams= pd.get_dummies(final_teams, prefix=['Team_1', 'Team_2'], columns = ['Team_1', 'Team_2'], dtype=int, sparse=False)
final_teams.head()
```

```python
X = final_teams.drop(columns=['winning_team'], axis=1)
y = final_teams['winning_team']
```

```python
# Splitting the data in the training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=34)
```

```python
X_train
```

```python
y_train
```

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, RMSprop
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Define the neural network model
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
# model.add(Dense(32, activation='selu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer=RMSprop(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1)

# Evaluate the model on the test set
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)

# Calculate accuracy and print classification report
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
fixtures = pd.read_csv("wc_prediction_data/fixtures.csv")
fixtures.head()
```

```python
pred_set=[]

mappings = {
    'Australia': 1, 'Pakistan': 2, 'India': 3, 'New Zealand': 4, 'England': 5, 'South Africa': 6, 'Bangladesh': 7,
    'Afghanistan': 8, 'Sri Lanka': 9, 'Netherlands': 10
}

# Insert team rankings into the fixtures dataframe based on 'Team_1' and 'Team_2' columns
fixtures.insert(1, 'team1_position', fixtures['Team_1'].map(mappings))
fixtures.insert(2, 'team2_position', fixtures['Team_2'].map(mappings))

fixtures = fixtures.iloc[:80, :]
fixtures.head()
```

```python
# Filling missing values in 'team1_position' and 'team2_position' columns with their respective median.

fixtures['team1_position']=fixtures['team1_position'].fillna(fixtures['team1_position'].median())
fixtures['team2_position']=fixtures['team2_position'].fillna(fixtures['team2_position'].median())
fixtures.head()
```

```python
# Predicting winning teams based on positions in fixtures.

for index, row in fixtures.iterrows():
  if row['team1_position'] < row['team2_position']:
    pred_set.append({'Team_1' :row['Team_1'] , 'Team_2' :row['Team_2'] , 'winning_team' : None })
  else:
    pred_set.append({'Team_1' :row['Team_2'] , 'Team_2' :row['Team_1'] , 'winning_team' : None })

pred_set = pd.DataFrame(pred_set)

pred_set.head()
```

```python
or_pred_set = pred_set
```

```python
# Create dummy variables(one hot encoding)
pred_set = pd.get_dummies(pred_set, prefix=['Team_1', 'Team_2'], columns=['Team_1', 'Team_2'], dtype=int)

# Finding missing columns in the "final_teams" dataset compared to the transformed "pred_set".
missing_cols = set(final_teams.columns) - set(pred_set.columns)
# missing_cols
```

```python
pred_set.head()
```

```python
# Setting the missing columns in "pred_set" dataframe to 0 and keeping only columns present in the "final_teams" DataFrame.
for cols in missing_cols :
  pred_set[cols] = 0

pred_set = pred_set[final_teams.columns]

pred_set = pred_set.drop(['winning_team'], axis=1)
pred_set.head()
```

```python
# Predict the single match aug_match_data of future

def predict_single_match(model, rankings, team_1, team_2):
  single_match_data = pd.DataFrame({
      'Team_1': [team_1],
      'Team_2':[team_2]
  })

  #Insert the team ranking data
  single_match_data.insert(1, 'team1_position',single_match_data['Team_1'].map(mappings))
  single_match_data.insert(2, 'team2_position',single_match_data['Team_2'].map(mappings))

  # Apply one hot encoding
  single_match_data = pd.get_dummies(single_match_data, prefix = ['Team_1', 'Team_2'], columns=['Team_1', 'Team_2'], dtype=int)

  #Find the missing columns
  missing_cols = set(final_teams.columns) - set(single_match_data.columns)

  #Set the missing columns to 0 and then keep only columns present in "final_teams"
  for col in missing_cols:
    single_match_data[col] = 0

  single_match_data = single_match_data[final_teams.columns]

  #Drop the winning column
  single_match_data = single_match_data.drop(['winning_team'], axis=1)

  #Making the prediction
  prediction = model.predict(single_match_data)
  prediction = (prediction > 0.5).astype(int)

  #print the result
  print(f"{team_1} vs {team_2}")

  if prediction[0] == 0 :
    print(f"Winner: {team_1}")
  else:
    print(f"Winner: {team_2}")

  print("************************")
```
