## Data transformations

## Building Nueral Network model Using Keras

import neccessaries libreries.

Here We use 2 dense layers with 64 nuerons in each layers and activation function as "RELU" and at output layer we use "softmax" as activation function.

```python
from keras.src.layers.serialization import activation
from keras.api._v2.keras.layers import Dense
from keras.src.engine.input_layer import Input
from tensorflow import keras
from keras.models import Sequential


inputs = Input(shape=(5,))

model = Sequential([
    keras.layers.InputLayer(input_shape=(5,)),
    keras.layers.Normalization(),
    keras.layers.Dense(64,activation="relu"),
    keras.layers.Dense(64,activation="relu"),
    keras.layers.Dense(1,activation = "softmax")
])
```

Compile the network use accuracy as matrix

```python
from keras.src.optimizers import RMSprop
model.compile(optimizer=RMSprop(learning_rate=0.001),loss="binary_crossentropy",metrics=["accuracy"])
```

fit the model on X

```python
model.fit(X,y,epochs=40,verbose=0)
```
